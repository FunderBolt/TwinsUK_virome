```python
%%bash
cut -f17 ../48MZ_Virome_Sample_Information.csv | tail -n +2 > samplesSequencingSpecificName.txt
filter_samples_from_otu_table.py -i TwinsUK_Batch3_OTU_table_w_tax.biom -o filtered_otu_table.biom --sample_id_fp samplesSequencingSpecificName.txt
biom convert -i filtered_otu_table.biom -o TwinsUK_DisCon.csv --to-tsv --header-key taxonomy
tail -n +2 TwinsUK_DisCon.csv > TwinsUK_DisCon.csv.mod
mv TwinsUK_DisCon.csv.mod TwinsUK_DisCon.csv
```

```python
mapping = pd.read_csv("../48MZ_Virome_Sample_Information.csv",sep='\t',header=0)
newLabels = dict(zip(mapping["SequencingSpecificName"],mapping["Sample_ID"]))
mapping.head()
```

```python
abundanceMatrix = pd.read_csv("TwinsUK_DisCon.csv",sep="\t")
taxonomy = dict(zip(abundanceMatrix["#OTU ID"],abundanceMatrix["taxonomy"]))
abundanceMatrix.set_index(["#OTU ID"],inplace=True)
del abundanceMatrix["taxonomy"]
abundanceMatrix.rename(columns=newLabels,inplace=True)
abundanceMatrix.head()
```

```python
abundanceMatrix.to_csv("matrix.bacteriaOTUs.csv",sep="\t",header=True,index=True)
```

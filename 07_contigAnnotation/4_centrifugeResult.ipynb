```python
import pandas as pd
import ete3
from ete3 import NCBITaxa 
ncbi = NCBITaxa()
```

```python
IN_DIR = "/ebio/abt3_projects/TwinsUK_viromes_Shao_Pei/tmp2/"
OUT_DIR = "/ebio/abt3_projects/TwinsUK_viromes_Shao_Pei/tmp2/0X_contigAnnotation/"
```

```python
#get centrifuge output
annotation = pd.read_csv(IN_DIR+"/0X_contigAnnotation/contigs.ForAlpha.centrifuge.viral.tsv",sep='\t',header=0)
annotation = annotation[annotation["seqID"]!="unclassified"]
annotation.head()
```

```python
report = pd.read_csv(IN_DIR+"/0X_contigAnnotation/contigs.ForAlpha.centrifuge_report.viral.tsv",sep='\t',header=0)
report.head()
```

```python
def getLineage(taxID):
    lineage = ncbi.get_lineage(taxID)
    names = ncbi.get_taxid_translator(lineage)
    return [names[taxid] for taxid in lineage]
```

```python
#build a dictionary of the annotated contigs
cVStax = annotation.merge(report,how="inner",left_on="taxID",right_on="taxID")
cVStax = cVStax.sort_values(by=["readID","score"],ascending=False)
cVStax["LINEAGE"] = cVStax["taxID"].apply(getLineage)
#save the annotations in a file
fw = open(OUT_DIR+"contigsVsVIRALtax.centrifuge.csv2",'w')
for index,row in cVStax.iterrows():
    print >> fw, row["readID"] + '\t' + ";".join(row["LINEAGE"])
fw.close()
```

```python
#Write only the best annoataiton for each contig. 
annotated = {}
for index,row in cVStax.iterrows():
    if row["readID"] in annotated:#In the future I should add all passible options.
        continue
    else:
        #annotated[row["readID"]] = [row["LINEAGE"],row["score"],(float(row["hitLength"])/row["queryLength"])]
        annotated[row["readID"]] = "; ".join(row["LINEAGE"])
        
fw = open(OUT_DIR+"contigsVsVIRALtax.centrifuge.csv",'w')
for contig in annotated:
    print >> fw, contig + '\t' + str(annotated[contig])
fw.close()
```

```python
from os import listdir
import pandas as pd
import numpy as np
```

```python
#Import bacteria references lengths file
referenceLengths =  pd.read_csv("0.bacterialGenomes/BacterialGenomes.length.txt",sep='\t',header=None)
referenceLengths.columns = ["REFERENCE","LENGTH"]
referenceLengths.set_index("REFERENCE",inplace=True)
referenceLengths.sort_values(by="LENGTH",inplace=True)
referenceLengths.head()
```

```python
#Import number of reads mapping per sample file
readsMapping = pd.read_csv("numReadsMapping.csv",header=0,index_col=0,sep='\t')
readsMapping.head()
```

```python
#Import coverage matrix per sample
#That is a matrix of bacteria contaminants X samples
folder = "7.coverageMatrices.sorted.bacteria/"
covSum = []
samples = []
for file in listdir(folder):
    if "coverage.matrix.csv" in file:
        samples.append(file.split('.')[0])
        covMatrix = pd.read_csv(folder+file,header=0,index_col=0,sep='\t')
        covMatrixSum = covMatrix.sum(axis=1)
        covSum.append(covMatrixSum)
        
        
        #Save Descriptive stats
        #stats = pd.DataFrame([covMatrixNormalized.median(axis=1),covMatrixNormalized.mean(axis=1)],index=["Median","Mean"])
        #stats = stats.T
        #stats.to_csv(matrix.split('.')[0]+".stats.csv",header=True,index=True,sep='\t')
```

```python
#Put all coverage matrices in a single one. 
#Samples without matching certain bacteria contaminant appears as 
covSum = pd.DataFrame(covSum)
covSum.index = samples
covSum = covSum*1000000000
```

```python
folder = "7.coverageMatrices.sorted.bacteria/"
covMatrix = pd.read_csv(folder+"Ab1.all.coverage.matrix.csv",header=0,index_col=0,sep='\t')
covMatrixSum = covMatrix.sum(axis=1)
```

```python
readsXgenome = []
for index in readsMapping.index.values:
    readsXgenome.append((referenceLengths*readsMapping.loc[index][0])["LENGTH"])
readsXgenome = pd.DataFrame(readsXgenome)
readsXgenome.index = readsMapping.index
readsXgenome.head()
```

```python
readsXgenome = readsXgenome[covSum.columns]
```

```python
covRPKM = covSum/readsXgenome
```

### Import bin stats

```python
folder = "7.coverageMatrices.sorted.bacteria/"
statMatrices = []
samples = []
for file in listdir(folder):
    if ".stats.csv" in file:
        samples.append(file.split('.')[0])
        statMatrix = pd.read_csv(folder+file,header=0,index_col=0,sep='\t')
        statMatrix = statMatrix["Median"]
        statMatrices.append(statMatrix)
statMatrices = pd.DataFrame(statMatrices)
statMatrices.index = samples
```

```python
statMatrices = statMatrices[contaminants]
```

```python
#Sort matrix from the most prevalent to the least prevalent (left to right)
statMatrices = statMatrices[(statMatrices>0).sum(axis=0).sort_values(ascending=False).index]
```

```python
statMatrices.to_csv("medianBinCoverage.allSamples.logNormalized.csv",header=True,index=True,sep='\t')
```
